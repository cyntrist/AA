Jupiter Notebook is a tool that allows us, as you can see, to switch between code and text. If you are not familiar with Markdown, it is a very simple text tagging standard that helps you build simple documents easily. The variables in a Notebook are not private to the cells but are shared, so you must be very careful when running the notebook, as the persistence of the variables is on the server to which the Notebook is connected.
This means that you can execute a cell in the middle of a notebook page, but you cannot guarantee that the value of the variable will be the one expected at that point in the code.
This sometimes causes strange errors. Before you go crazy trying to find the fault, try resetting the kernel. To do this, you can go to **Kernel/Reset Kernel** or you can press the double black arrow button (pink box in Figure 1).

The results of the last Python instruction are displayed below the cell. If there are previous return values, they are deleted unless they are explicitly displayed from Python. One way to display this is by using the Python print() method, but there is another method called **display()**. The display() method is unique to Jupiter and can be used within Python to display result information anywhere in the code.
It is similar to print, but with some data types, such as dataframes, it formats them visually, making it easier to read and understand the result obtained.
